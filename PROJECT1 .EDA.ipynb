## Average Price per flight in route

```python
data_1['Price'] = data_1['Price'].str.replace(',', '').astype(float)
data1_avg_price = data_1.groupby('FlightName')['Price'].mean().reset_index().astype({'Price':'int'}).sort_values(by='Price',ascending=False)
p = sns.barplot(x='FlightName',y='Price',data=data1_avg_price)
# p.bar_label(p.containers[0])
plt.title("Average Price per flight in route")
p.set_xticklabels(labels=data1_avg_price["FlightName"], rotation=35)
sns.set(font_scale=1.25)
plt.show()
```

```python
df_dh_avg = data_1.groupby('DepartingTime')['Price'].mean().reset_index()
df_dh_avg.plot.line(x='DepartingTime',y='Price',title='DepartingHour Vs Price')
plt.ylabel('Price')
plt.show()
```

# **CONCATENATING ALL ROUTES**

```python
df.shape
```

```python
df_avg = df.groupby('FlightName')['Price'].mean().reset_index().astype({'Price':'int'}).sort_values(by='Price',ascending=False)
df_avg
```

Average price per flight

```python
df_avg.plot.bar(x='FlightName',y='Price')
plt.show()
```

## OBSERVTION:
Vistara is the costliest airline in the country.

# AVERAGE PRICE PER ROUTE

```python
df_avg_route = df.groupby(['DepartingCity','ArrivingCity'])['Price'].mean().reset_index().astype({'Price':'int'}).sort_values(by='Price',ascending=False)
df_avg_route["Route"] = df_avg_route["DepartingCity"]+"↔️"+df_avg_route["ArrivingCity"]
df_avg_route
```

```python
df_avg_route.plot.bar(x='Route',y='Price')
plt.show()
```
